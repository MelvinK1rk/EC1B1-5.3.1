<font size="3">__Creating the initial dataframe and making it look nice__</font> 

Merging japan_data and usa_data into a new dataframe called <font color='red'>combined</font> 

```python
combined = pd.merge(left = japan_data, right = usa_data, left_index = True, right_index = True)
combined
```

Removing the duplicate date column

```python
combined.drop(['International Financial Statistics (IFS)_y'], axis = 1, inplace = True)
```

Renaming the column labels, removing the first two rows, formatting the month and year into a date format and making the dates the index in <font color='red'>combined</font> 

For Japan, rename:

* Economic Activity, Industrial Production, Index to ***production***
* Exchange Rates, National Currency Per U.S. Dol... to ***exchange_rate_yen_to_dollar***
* International Reserves and Liquidity, Reserves... to ***reserves_japan***
* Prices, Consumer Price Index, All items, Index to ***cpi_japan***

For the US, rename: 

* International Reserves and Liquidity, Reserves... to ***reserves_usa***
* Prices, Consumer Price Index, All items, Index to ***cpi_usa***

NB reserves are in millions of dollars

```python
combined.columns = ['date', 'production', 'exchange_rate_yen_to_dollar', 'reserves_japan', 'cpi_japan', 'reserves_usa', 'cpi_usa']
combined = combined.drop([0,1]).set_index(['date']).astype(float)
combined.index = pd.to_datetime(combined.index)
combined
```

<font size="3">__Constructing the various variables__</font> 

Constructing a variable, ***log_exchange_rate***, for the log exchange rate and creating a new column for it in <font color='red'>combined</font> 

```python
log_exchange_rate = np.log(combined['exchange_rate_yen_to_dollar'])
combined['log_exchange_rate'] = log_exchange_rate
combined
```

Constructing a variable, ***log_exchange_rate_diff***, for the monthly difference in ***log_exchange_rate*** and creating a new column for it in <font color='red'>combined</font>

```python
log_exchange_rate_diff = combined['log_exchange_rate'].diff()
log_exchange_rate_diff[0] = 0.00 #Set log_exchange_rate_diff during Jan 1960 to 0 as Japan still part of Bretton Woods hence fixed exchange rates
combined['log_exchange_rate_diff'] = log_exchange_rate_diff
combined
```

Constructing a variable, ***inflation_japan***, for monthly inflation in Japan and creating a new column for it in <font color='red'>combined</font>.

Note that monthly inflation is (approximately) equal to the monthly difference in the log of ***cpi_japan***

NB: monthly inflation will not be in percentages

```python
inflation_japan = np.log(combined['cpi_japan']).diff()
inflation_japan[0] = inflation_japan[1] #set Japan's inflation for Jan 1960 to Feb 1960's value
combined['inflation_japan'] = inflation_japan
combined
```

Constructing a variable, ***log_production***, for the log of industrial production in Japan and creating a new column for it in <font color='red'>combined</font>.

```python
log_industrial_production = np.log(combined['production'])
combined['log_production'] = log_industrial_production
combined
```

Constructing a variable, ***log_production_diff_month***, for the monthly difference (growth) of the log of industrial production in Japan and creating a new column for it in <font color='red'>combined</font>.

```python
log_production_diff_month = combined['log_production'].diff()
log_production_diff_month[0] = log_production_diff_month[1] #set NaN value to Feb's 1960 value
combined['log_production_diff_month'] = log_production_diff_month
combined
```

Constructing a variable, ***log_production_diff_year***, for the yearly difference (growth) of the log of industrial production in Japan and creating a new column for it in <font color='red'>combined</font>.

```python
log_production_diff_year = combined['log_production'].diff(12)
combined['log_production_diff_year'] = log_production_diff_year
combined
```

Constructing a variable, ***reserves_index_japan***, for the value of Japan's reserves expressed in terms of the value in January 1960 and creating a new column for it in <font color='red'>combined</font>.

```python
reserves_index_japan = 100 * combined['reserves_japan'] / combined.loc['1960-01-01', 'reserves_japan' ]
combined['reserves_index_japan'] = reserves_index_japan
combined
```

Constructing a variable, ***inflation_usa***, for monthly inflation in the USA and creating a new column for it in <font color='red'>combined</font>.

```python
inflation_usa = np.log(combined['cpi_usa']).diff()
inflation_usa[0] = inflation_usa[1] #set Japan's inflation for Jan 1960 to Feb 1960's value
combined['inflation_usa'] = inflation_usa
combined
```

Constructing a variable, ***reserves_index_usa***, for the value of the USA's reserves expressed in terms of the value in January 1960 and creating a new column for it in <font color='red'>combined</font>.

```python
reserves_index_usa = 100 * combined['reserves_usa'] / combined.loc['1960-01-01', 'reserves_usa' ]
combined['reserves_index_usa'] = reserves_index_usa
combined
```

<font size="3">__Checking for outliers__</font> 

* Interquartile range = ***iqr***

An outlier is defined to be a value which is 1.5***iqr*** above or below the lower or upper quartile.

We will check all columns to see how many values are outliers and do further checks to determine if they should be removed.

Below, we count how many outliers there are.

```python
describe = combined.describe()
describe
```

```python
iqr = describe.loc['75%'] - describe.loc['25%']
iqr
```

Values above ***upper_threshold*** or below ***lower_threshold*** are outliers

```python
upper_threshold = 1.5 * iqr + describe.loc['75%']
lower_threshold = describe.loc['25%'] - 1.5 * iqr

print(upper_threshold, lower_threshold)
```

Creating a boolean mask that sets a value in <font color='red'>combined</font> to True if it is between ***upper_threshold*** and ***lower_threshold*** (ie not an outlier), then applying said mask to <font color='red'>combined</font>

```python
not_outlier = (combined > lower_threshold) & (combined < upper_threshold)
not_outlier
```

Counting how many False values in the boolean mask (ie how much outliers there are)

```python
(~not_outlier).sum()
```

There are an unreasonable amount of outliers in ***reserves_japan***, ***log_exchange_rate_diff***, ***reserves_index_japan***, ***reserves_usa***, ***log_production_diff_year***, ***reserves_index_usa***, ***inflation_usa*** and ***inflation_japan***  so we tighten the criteria for identifying outliers to 2***iqr*** above or below the upper or lower quartile


Repeating what we did above

```python
upper_threshold_2 = 2 * iqr + describe.loc['75%']
lower_threshold_2 = describe.loc['25%'] - 2 * iqr

print(upper_threshold_2, lower_threshold_2)
```

```python
not_outlier_2 = (combined > lower_threshold_2) & (combined < upper_threshold_2)
not_outlier_2
```

```python
(~not_outlier_2).sum()
```

Creating a seaborn heatmap to visualise where all the outliers are. The cream parts are where null values are (for the majority of null values, they are outliers)

```python
sns.heatmap(combined[not_outlier_2].isnull(), cbar = False)
plt.show()
```

Some columns still have too much outliers so we create graphs to identify any trends in the data **(NOT FOR SECTION 5)**

```python
for column in combined.columns:
    plt.plot(combined[column])
    plt.title(column)
    plt.show()
```

The graph for ***reserves_usa*** and ***reserves_index_usa*** shows a sharp upward trend in the value of the USA's reserves past 1988. The heatmap shows all the outliers are for values past 1988. We shall not remove any outliers for these columns. 

For a similar reason, we shall not remove any outliers for ***reserves_japan*** and ***reserves_index_japan***.

Regarding ***log_exchange_rate_diff***, the outliers occur after Japan left Bretton Woods. We cannot remove several years worth of data and it is unlikely the International Monetary Fund screwed up in collecting these values. Also, it is plausible wild fluctuations in ***log_exchange_rate_diff*** did occur. We shall not remove any outliers.

Considering the fact Japan's inflation rate peaked at less than 25% during the 1970s, an inflation rate of 40% seems unlikely. We shall remove the outliers in ***inflation_japan***.

There are no outliers in ***log_production***. As ***log_production_diff_month*** and ***log_production_diff_year*** follow on from ***log_production***, we can't remove any outliers in these columns. The outliers represent the decrease in industrial production in Japan during the 1970s. 

We will replace all the outliers in ***inflation_usa***.



Setting all outliers to NaN

```python
combined.loc[~not_outlier_2['inflation_japan'], 'inflation_japan'] = np.nan
combined.loc[~not_outlier_2['inflation_usa'], 'inflation_usa'] = np.nan
```

<font size="3">__Constructing a series for monthly real exchange rate growth__</font> 

Constructing a variable, ***monthly_real_exchange_rate_growth***, for monthly real exchange rate growth (yen to dollar) and creating a new column for it in <font color='red'>combined</font>.

***monthly_real_exchange_rate_growth*** = ***log_exchange_rate_diff*** + ***inflation_japan*** - ***inflation_usa***
